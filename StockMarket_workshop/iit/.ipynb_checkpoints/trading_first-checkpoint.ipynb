```python
# Imports
import pandas as pd
import talib as ta
import matplotlib.pyplot as plt

%matplotlib inline
```

#### Download trading data

#### Load data

```python
df = pd.read_csv('yesbank.csv', index_col='Date')
df.keys()
df
```

##### Line ploting of closing price

```python
df[['Close Price']].plot(figsize=(15,10))
```

```python
df['Close Price SMA 30'] = ta.SMA(df['Close Price'])
df['Close Price SMA 20'] = ta.SMA(df['Close Price'], timeperiod=20)
df['Close Price SMA 10'] = ta.SMA(df['Close Price'], timeperiod=10)
df['Close Price SMA 5'] = ta.SMA(df['Close Price'], timeperiod=5)
```

```python
df[['Low Price SMA 5', 'Low Price']].plot(
    figsize=(15,10),
    grid=True,
    legend=True
)
```

#### Calculate Low price EMA and WMA

```python
df['Low Price EMA 30'] = ta.EMA(df['Low Price'])
df['Low Price EMA 20'] = ta.EMA(df['Low Price'], timeperiod=20)
df['Low Price EMA 10'] = ta.EMA(df['Low Price'], timeperiod=10)
df['Low Price EMA 5'] = ta.EMA(df['Low Price'], timeperiod=5)
```

```python
df['Low Price WMA 30'] = ta.WMA(df['Low Price'])
df['Low Price WMA 20'] = ta.WMA(df['Low Price'], timeperiod=20)
df['Low Price WMA 10'] = ta.WMA(df['Low Price'], timeperiod=10)
df['Low Price WMA 5'] = ta.WMA(df['Low Price'], timeperiod=5)
```

```python
df[['Low Price WMA 10', 'Low Price SMA 10', 'Low Price EMA 10', 'Low Price']].plot(
    figsize=(15,10),
    grid=True, 
    legend=True
)
```
